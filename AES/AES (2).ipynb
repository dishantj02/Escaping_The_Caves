```python
to_cipher = {'0':'f', '1':'g', '2':'h', '3':'i', '4':'j', '5':'k', '6':'l', '7':'m', '8':'n', '9':'o', 'a':'p', 'b':'q', 'c':'r', 'd':'s', 'e':'t', 'f':'u'}
to_hex = {'f':'0', 'g':'1', 'h':'2', 'i':'3', 'j':'4', 'k':'5', 'l':'6', 'm':'7', 'n':'8', 'o':'9', 'p':'a', 'q':'b', 'r':'c', 's':'d', 't':'e', 'u':'f'}

def cipher_to_hex(word):
    new=''
    for i in word:
        new+=to_hex[i]
    return new

def hex_to_cipher(word):
    new=''
    for i in word:
        new+=to_cipher[i]
    return new
```
